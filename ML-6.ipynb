```python
import numpy as np
import pandas as pd

#read the data source file
train_set = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', header = None)
```

```python
train_set.head()
```

```python
#change the coumn names
train_set.columns =['age', 'workclass', 'fnlwgt', 'education','education_num', 'marital_status','occupation', 'relationship','race', 'sex','capital_gain', 'capital_loss','hours_per_week', 'native_country', 'salary_slab']
```

```python
#check for the top rows
train_set.head()
```

```python
#check salary slab count
train_set.salary_slab.value_counts()
```

```python
#analyse the data
train_set.describe()
```

```python
#replace ? with NaN
dataset = train_set.replace('?',np.NaN)
dataset.shape
```

```python
#replace null values with 0
dataset.fillna(0, inplace=True)
```

```python
X = dataset[['age', 'workclass', 'fnlwgt', 'education', 'education_num',
       'marital_status', 'occupation', 'relationship', 'race', 'sex',
       'capital_gain', 'capital_loss', 'hours_per_week', 'native_country']]
Y = dataset['salary_slab']
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
#label encode all the categorical values
labelencoder_sex = LabelEncoder()
X.iloc[:,9] = labelencoder_sex.fit_transform(X.iloc[:,9])

labelencoder_race = LabelEncoder()
X.iloc[:,8] = labelencoder_race.fit_transform(X.iloc[:,8])

labelencoder_country = LabelEncoder()
X.iloc[:,13] = labelencoder_country.fit_transform(X.iloc[:,13])

labelencoder_relation = LabelEncoder()
X.iloc[:, 7] = labelencoder_relation.fit_transform(X.iloc[:, 7])

labelencoder_occupation = LabelEncoder()
X.iloc[:, 6] = labelencoder_occupation.fit_transform(X.iloc[:, 6])

labelencoder_marital = LabelEncoder()
X.iloc[:, 5] = labelencoder_marital.fit_transform(X.iloc[:, 5])

labelencoder_education = LabelEncoder()
X.iloc[:, 3] = labelencoder_education.fit_transform(X.iloc[:, 3])

labelencoder_workclass = LabelEncoder()
X.iloc[:, 1] = labelencoder_workclass.fit_transform(X.iloc[:, 1])
```

```python
#hotencode multivalues attributes
onehotencoder = OneHotEncoder(categorical_features=[1, 3, 5, 6, 7, 8, 13])
X = onehotencoder.fit_transform(X).toarray()
```

```python
#convert ndarray to dataframe
X = pd.DataFrame(X)
X.head()
```

```python
from sklearn.model_selection import train_test_split

#split into train and test sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
from xgboost import XGBClassifier

#create train and test data
classifier = XGBClassifier()
classifier.fit(X_train, Y_train)
```

```python
#predict the test data with created model
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
#check right predictions in confusion matix
cm = confusion_matrix(Y_test, y_pred)
cm
```

```python
#efficiency of the model will be
(cm[1][1]+cm[0][0])/(cm[0][1]+cm[0][0]+cm[1][0]+cm[1][1])
```
